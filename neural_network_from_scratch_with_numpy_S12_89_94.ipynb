   - L1正则化的正则化项为模型参数的绝对值和。

### 例子

其中：
- \(\text{MSE}\) 是均方误差。
- \(\lambda\) 是正则化参数。
- \(w_i\) 是模型的参数。

### Softmax 函数

#### 为什么要用 `exp_scores = np.exp(a2)` 和 `softmax = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)`？

Softmax 函数通常用于多分类问题的输出层，它将网络的输出转换为概率分布。Softmax 函数的公式如下：

\[ \text{softmax}(z_i) = \frac{e^{z_i}}{\sum_{j} e^{z_j}} \]

其中 \( z_i \) 是输出层的第 \( i \) 个元素。

1. **`exp_scores = np.exp(a2)`**：
   - 这行代码计算每个输出元素的指数。指数函数 \( e^{z_i} \) 可以将输出值转换为正数，并且可以放大差异。

2. **`softmax = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)`**：
   - 这行代码计算每个输出元素的 softmax 值。通过将每个指数值除以所有指数值的总和，可以将输出值转换为概率分布。`axis=1` 表示在每一行上进行求和，`keepdims=True` 保持输出的维度。

### 损失函数

#### 为什么损失函数要这样写？

损失函数用于衡量模型的预测值与真实值之间的差异。在分类问题中，常用的损失函数是交叉熵损失（Cross-Entropy Loss）。此外，还可以添加正则化项来防止过拟合。

```python
def loss(softmax: np.ndarray, y: np.ndarray, model_dic: dict) -> float:

    total_loss = softmax_loss + reg_loss
    return total_loss / y.shape[0]
```

1. **`W1, b1, W2, b2 = retrieve(model_dic)`**：
   - 从字典中检索模型参数。

2. **`m = np.zeros(200)`**：
   - 初始化一个数组 `m`，用于存储每个样本的预测概率。

3. **`for i, correct_index in enumerate(y): predicted = softmax[i][correct_index] m[i] = predicted`**：
   - 遍历每个样本，获取其对应的预测概率，并存储在数组 `m` 中。

4. **`log_prob = -np.log(m)`**：
   - 计算每个样本的对数概率。对数概率用于计算交叉熵损失。

5. **`softmax_loss = np.sum(log_prob)`**：
   - 计算所有样本的 softmax 损失。

6. **`reg_loss = lambda_reg / 2 * (np.sum(np.square(W1)) + np.sum(np.square(W2)))`**：
   - 计算正则化损失。正则化项用于防止过拟合，通过惩罚模型参数的大小来实现。

7. **`total_loss = softmax_loss + reg_loss`**：
   - 计算总损失，即 softmax 损失和正则化损失的和。

8. **`return total_loss / y.shape[0]`**：
   - 返回平均损失，即总损失除以样本数量。
