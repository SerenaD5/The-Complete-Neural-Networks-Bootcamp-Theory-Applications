###  uniform initialization
uniform distribution

image.png

image.png

And this underscore here is just an in-place operation. So you don't have to define this and set it equal to your tensor.
You can directly perform this function on your tensor directly.

```python
# 1st, uniform initialization
# nn.init.uniform_(tensor, a=0.0, b=1.0, generator=None)
nn.init.uniform_(layer.weight.data, a=0.0, b=3)
```

```python
nn.init.uniform_(layer.weight, a=0.0, b=3)
# And you can also remove this, it's not a problem.
# If you remove this, it will just tell you that this is a parameter and whether it requires grad or not.
# So in this case, since this is a weight,that means it does require grad,does require automatic differentiation,it does require back propagation.
```

### normal initialization

image.png

you initialize a tensor with a normal distribution or a gouging distribution as we learned, that has a mean of 0 and the standard deviation of 1.
So this is the typical one, a mean of 0 and standard deviation of 1. If you'd like to change it, you can directly change the mean value and the standard deviation value.

```python
#nn.init.normal_(tensor, mean=0.0, std=1.0, generator=None)
nn.init.normal_(layer.weight, mean=0.0, std=1.0)
```

```python
nn.init.normal_(layer.weight, mean=0.0, std=0.2)
# 
```

###  Xavier initialization

image.png

So if you remember from the theory section,the Xavier initialization initialize your weight tensor from a uniform distribution.
But the only difference is that the uniform distribution is basically specified by a specific value,which is calculated depending on the number of neurons at your inputs and number of neurons at your outputs.
So in this case, you first calculate this value,which is the number of neurons at your inputs,number of neurons at your output.So in this case, you first calculate this value,which is the number of neurons at your inputs,number of neurons at your output.Then this formula, you get A.And then you initialize your weight from the uniform distribution and the minimum value is -A and A. So A is calculated depending on this formula right here.So this is in the case of a uniform distribution.
And of course you can also do it in the case of a normal distribution.In this case, you initialize it from the normal distribution with a mean of 0 and a standard deviation,which you're going to be calculating here,again, depending on the number of input neurons and the number of output neurons.
